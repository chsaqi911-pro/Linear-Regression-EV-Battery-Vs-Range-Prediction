```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("/content/drive/MyDrive/EV Dataset.xlsx")
```

```python
print(df)
```

```python
plt.scatter(df.battery_capacity_ah, df.range_km)
```

```python
plt.xlabel('battery_capacity_ah', color="Blue")
plt.ylabel('range_km', color = "blue")
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['battery_capacity_ah']], df.range_km)
```

```python
lr.predict([[75]])
```

```python
lr.score(df[['battery_capacity_ah']], df.range_km)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
print (((5.486)*75) + -72.859)
```

```python
plt.xlabel('battery_capacity_ah', color="Blue")
plt.ylabel('range_km', color = "blue")
plt.title('EV Battery Capacity Vs Range Prediction Model')
plt.scatter(df.battery_capacity_ah, df.range_km)
plt.plot(df.battery_capacity_ah, lr.predict(df[['battery_capacity_ah']]), color = 'red')
```
